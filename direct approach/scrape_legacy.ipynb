```python
import pdfplumber

# Open the file and create a pdf object.
pdf = pdfplumber.open("llamasoft-llc_data-science-intern.pdf")

# Get the number of pages.
numPages = len(pdf.pages)

print("Number of Pages:", numPages)

# Iterate over each page and extract the text of each page.
flag = 0
text = ""
for number, pageText in enumerate(pdf.pages):
    print("Page Number:", number)
    if "Program" in pageText.extract_text() or flag == 1:
        text += (pageText.extract_text())
        flag = 1
# print(text)
lines = text.split('\n')

flag = 0
filtered_lines = []
for line in lines:
    if "Program" in line:
        flag = 1
    if "MSR" in line:
        filtered_lines.append(line)
        break
    if "Eligibilty" in line:
        continue
    if flag == 1:
        filtered_lines.append(line)
    else:
        continue

filtered_lines
```
