### Sample Code for a Single File:

```python
pdfFile = wi(filename = 'ab-inbev_technology-intern.pdf', resolution = 300)
image = pdfFile.convert('jpeg')
```

```python
imageBlobs = []

for img in image.sequence:
	imgPage = wi(image = img)
	imageBlobs.append(imgPage.make_blob('jpeg'))

extract = []

for imgBlob in imageBlobs:
	image = Image.open(io.BytesIO(imgBlob))
	text = pytesseract.image_to_string(image, lang = 'eng')
	extract.append(text)
```

```python
extractStr = ''.join(str(e) for e in extract)
print(extractStr)
```

```python
list_el = (extractStr[extractStr.find('Program'):extractStr.find('Msc')]).split('\n\n')
list_el[1] = (list_el[1])[:list_el[1].find('\n')]

list_el[0] = list_el[0].split(' ')
list_el[1] = list_el[1].split(' ')


for i,s in enumerate(list_el[1]):
    if(s == 'YesYes'):
        list_el[1][i] = 'Yes'
        (list_el[1]).insert(i, 'Yes')
    if(s == 'NoNo'):
        list_el[1][i] = 'No'
        (list_el[1]).insert(i, 'No')
    if(s == '--'):
        list_el[1][i] = 'No'

branchEl_map = {}
for i,s in enumerate(list_el[1]):
    if(i == 0):
        continue
    branchEl_map[list_el[0][i]] = list_el[1][i]

branchEl_map
```
